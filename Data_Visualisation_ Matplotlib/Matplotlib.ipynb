```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
#linspace (start, stop, num)- Return evenly spaced numbers over a specified interval.
X = np.linspace(0,5,11)
Y =  X**2
```

```python
X
```

```python
Y
```

```python
# Fuction approach
plt.plot(X,Y)
plt.xlabel('Profit')
plt.ylabel('Market')
plt.title('Global sales')
```

```python
#Object oriented method
fig = plt.figure()
#add_axes([left,bottom,width,height]) method figure module of matplotlib library is used to add an axes to the figure.
#
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(X,Y)
axes.set_xlabel('PROFIT')
```

```python
fig2 = plt.figure()
axes1 =fig2.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig2.add_axes([0.69,0.12,0.2,0.2])
axes1.plot(X,Y)
axes2.plot(X,Y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
#when number of rows are maby abd  axis overlappes use tight_layout
#Iterating axes
# for current_ax in axes:
#     current_ax.plot(X,Y)
axes[0].plot(X,Y)
axes[1].plot(Y,X)
axes[0].set_title('Profit')
axes[1].set_title('Sales')
plt.tight_layout()
```

**Figure size and DPI**

```python
fig3 = plt.figure()
axes3 = fig3.add_axes([0,0,1,1])
axes3.plot(X,X**2,label='X Squared')
axes3.plot(X,X**3,label='X Cubed')
#LOC PARAMETER 'best'	0
# 'upper right'	1
# 'upper left'	2
# 'lower left'	3
# 'lower right'	4
# 'right'	5
# 'center left'	6
# 'center right'	7
# 'lower center'	8
# 'upper center'	9
# 'center'	10
axes3.legend(loc=0)
```

**PLOT APPEARANCE**

```python
pfig = plt.figure()
pax = pfig.add_axes([0,0,1,1])
pax.plot(X,Y,Y,X,color='#144D5B',lw=2,ls='solid',marker='o')
pax.grid(True)
```
