```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,color='blue')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm',vars=['total_bill','size'])
```

```python
sns.rugplot(tips['tip'])
```

```python
sns.kdeplot(tips['total_bill'])
```

#Categorial plot

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxplot(y='total_bill',x='sex',data=tips,hue='smoker')
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.violinplot(y='total_bill',x='sex',data=tips,hue='day')
```

```python
sns.swarmplot(x='sex',y='tip',data=tips,hue='day',hue_order=['Sat','Sun','Thur','Friday'],dodge='sex')
```

```python
sns.swarmplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
```

#Matrix plots.

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tc = tips.corr()
sns.heatmap(tc,annot=True)
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linewidths=0.4,linecolor='white')
```

```python
sns.clustermap(fp,cmap='coolwarm')
```

#grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
p = sns.PairGrid(iris)
p.map_diag(sns.distplot)
p.map_upper(plt.scatter)
p.map_lower(sns.kdeplot)
```
